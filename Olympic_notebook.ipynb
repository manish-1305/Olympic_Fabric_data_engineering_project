```python
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, DoubleType, BooleanType, DateType
```

```python
athletes = spark.read.format("csv").option("header","true").option("inferSchema","true").load("Files/raw/Athletes.csv")
coaches = spark.read.format("csv").option("header","true").option("inferSchema","true").load("Files/raw/Coaches.csv")
entriesgender = spark.read.format("csv").option("header","true").option("inferSchema","true").load("Files/raw/EntriesGender.csv")
medals = spark.read.format("csv").option("header","true").option("inferSchema","true").load("Files/raw/Medals.csv")
teams = spark.read.format("csv").option("header","true").option("inferSchema","true").load("Files/raw/Teams.csv")
     
```

```python
athletes.show()
athletes.printSchema()
```

```python
coaches.show()
coaches.printSchema()
```

```python
entriesgender.show()
entriesgender.printSchema()
```

```python
medals.show()
medals.printSchema()
```

```python
medals = medals.withColumnRenamed("Rank by Total", "Rank_by_Total")
medals.show()
```

```python
teams.show()
teams.printSchema()
```

```python
# Find the top countries with the highest number of gold medals
top_gold_medal_countries = medals.orderBy("Gold", ascending=False).select("TeamCountry","Gold").show()
```

```python
# Calculate the average number of entries by gender for each discipline
average_entries_by_gender = entriesgender.withColumn(
    'Avg_Female', entriesgender['Female'] / entriesgender['Total']
).withColumn(
    'Avg_Male', entriesgender['Male'] / entriesgender['Total']
)
average_entries_by_gender.show()
     
```

**Loading the data to the destination folder**

```python
athletes.repartition(1).write.mode("overwrite").option("header",'true').csv("Files/transformed-data/athletes")
```

```python
coaches.repartition(1).write.mode("overwrite").option("header","true").csv("Files/transformed-data/coaches")
entriesgender.repartition(1).write.mode("overwrite").option("header","true").csv("Files/transformed-data/entriesgender")
medals.repartition(1).write.mode("overwrite").option("header","true").csv("Files/transformed-data/medals")
teams.repartition(1).write.mode("overwrite").option("header","true").csv("Files/transformed-data/teams")
     
```
